@Title : Program Aim is to create bar plots with error bars on the same figure.

```python
import matplotlib.pyplot as plt
from LoggerFormat import logger
import numpy as np

try:
    mean_velocity = (54.74, 42.35, 67.37, 58.24, 30.25)
    sd_of_velocity = (4, 3, 4, 1, 5)
    index = np.arange(len(mean_velocity))
    fig,ax = plt.subplots()
    rects1 = ax.bar(index,mean_velocity,width=0.35,yerr = sd_of_velocity)

    plt.xlabel("Velocity")
    plt.ylabel("scores")
    plt.title("Scores by Velocity")

    plt.minorticks_on()
    plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
    plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
    plt.show()

except Exception as err:
    logger.error(err)
```
