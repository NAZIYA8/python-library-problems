@Title : Program Aim is to display a bar chart of the popularity of programming
         Languages.Attach a text label above each bar displaying its popularity (float value).

```python
import matplotlib.pyplot as plt
from LoggerFormat import logger

try:
    lang = ['Java', 'Python','PHP', 'JavaScript', 'C#', 'C++']
    popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    plt.bar(lang,popularity)
    plt.xlabel("Programming languages")
    plt.ylabel("Popularity")
    plt.title("Popularity of programming languages")

    for i in range(len(lang)):
        plt.text(i,popularity[i]+0.5,popularity[i],ha='center')
    plt.ylim(0,25)    
    plt.show()

except Exception as err:
    logger.error(err)
```
