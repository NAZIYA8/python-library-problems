@Title : Program Aim is to scatter plot using random distributions to generate balls of
         different sizes.

```python
import matplotlib.pyplot as plt
from LoggerFormat import logger
import numpy as np
import math
import random

try:
    x = np.random.randint(0,100,20)
    y = np.random.randint(0,100,20)
    plt.figure(figsize=(16,6))
    colors = [np.random.randint(1, 4) for i in range(20)]
    areas = [math.pi * random.randint(5, 15)**2 for i in range(20)]
    plt.scatter(x, y, s=areas, c= colors, alpha = 0.7)
    
    plt.xlabel("X")
    plt.ylabel("Y")

except Exception as err:
    logger.error(err)
```
