@Title : Program Aim is to draw a scatter plot to find sea level rise in past 100 years.

```python
import matplotlib.pyplot as plt
import pandas as pd

try:
    data = pd.read_csv('epa-sea-level_csv.csv')
    year = data['Year']
    sea_levels = data['CSIRO Adjusted Sea Level']
    plt.figure(figsize=(16,8))
    plt.scatter(year, sea_levels, c='g')
    plt.xlabel('Year')
    plt.ylabel('Sea Level')
    plt.title('Rise in Sealevel')
    plt.show()

except Exception as err:
    logger.error(err)
```
