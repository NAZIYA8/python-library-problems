@Title : Program Aim is to plot two or more lines on same plot with suitable legends of each
         line.

```python
from matplotlib import pyplot as plt
from LoggerFormat import logger

try:
    x1 = [10,20,30]
    y1 = [20,40,10]
    plt.plot(x1, y1, label = "Line 1")

    x2 = [10,20,30]
    y2 = [40,10,30]
    plt.plot(x2, y2, label = "Line 2")

    plt.xlabel('x - axis')
    plt.ylabel('y - axis')
    plt.title('Two or more lines on same plot with suitable legends ')
    plt.legend()
    plt.show()
except Exception as err:
    logger.error(err)
```
