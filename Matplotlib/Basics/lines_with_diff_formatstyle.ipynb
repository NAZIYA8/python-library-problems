         using arrays.

```python
import numpy as np
import matplotlib.pyplot as plt
from LoggerFormat import logger 

try:
    time = np.arange(0., 5., 0.2)
    plt.plot(time, time, 'b^', time, time**2, 'ks', time, time**3, 'r--')
    plt.show()

except Exception as err:
    logger.error(err)
```
