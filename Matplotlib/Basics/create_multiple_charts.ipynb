```python

import matplotlib.pyplot as plt
from LoggerFormat import logger

try:
    groups = ["A","B","C"]
    values = [20,50,90]
    plt.figure(figsize=(16,6))

    plt.subplot(1,3,1)
    plt.plot(groups,values,color="g")

    plt.subplot(1,3,2)
    plt.bar(groups,values,color="b")

    plt.subplot(1,3,3)
    plt.scatter(groups,values,marker="*")

except Exception as err:
    logger.error(err)
```
